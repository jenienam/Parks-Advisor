```python
import pandas as pd
import glob
import os
import numpy as np
from sqlalchemy import create_engine

path = r'C:\Course Work\ETL Project\WildlifeTables'
all_files = glob.glob(os.path.join(path, "*.csv"))

df_from_each_file = (pd.read_csv(f) for f in all_files)
species_df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
species_df.head()
```

```python
species_df.columns
```

```python
clean_species_df = species_df[['Park Code', 'Park Name', 'Category', 'Order', 'Family', 'Scientific Name', 'Common Names', 'Synonyms', 'Occurrence', 'Nativeness', 'Abundance', 'NPS Tags']]
clean_species_df.head()
```

```python
clean_species_df.loc[:, "Synonyms"].fillna("None", inplace = True)
clean_species_df.loc[:, "Abundance"].fillna("Unknown", inplace = True)
clean_species_df.loc[:, "NPS Tags"].fillna("None", inplace = True)
clean_species_df.loc[:, "Order"].fillna("Unknown", inplace = True)
clean_species_df.loc[:, "Family"].fillna("Unknown", inplace = True)
clean_species_df.loc[:, "Common Names"].fillna("Unknown", inplace = True)
clean_species_df.loc[:, "Nativeness"].fillna("Unknown", inplace=True)

clean_species_df.head()
```

```python
clean_species_df = clean_species_df[clean_species_df['Occurrence'].notna()]
clean_species_df.head()
```

```python
clean_species_df.index = np.arange(1, len(clean_species_df) + 1)
clean_species_df.head()
```

```python
clean_species_df.Category.unique()
```

```python
wildlife_df = clean_species_df[clean_species_df['Category'].isin(['Mammal', 'Bird', 'Reptile', 'Amphibian', 'Fish'])]
wildlife_df.head()
```

```python
wildlife_df.to_csv("clean_wildlife.csv", index=False)
```
